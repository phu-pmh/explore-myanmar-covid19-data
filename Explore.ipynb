## တစ်ရက်ချင်းစီအတွက် data ကိုကြည့်ရအောင်

```python
import datetime
from datetime import date
import re

daily_data=[]
for file in filenames:
    match = re.search(r'\d{2}-\d{2}-\d{4}', file)
    date = datetime.datetime.strptime(match.group(), '%d-%m-%Y').date()
    tmp_df = pd.read_csv(file).sum()
    tmp_df['Date'] = date
    daily_data.append(tmp_df)
cols = ['Date']+df_sum_by_region.columns.tolist()
```

```python
df_daily_data = pd.DataFrame(daily_data)[cols].sort_values(by=['Date']).fillna(0)
df_daily_data
```

## Cumulative Confirmed Plot (စုစုပေါင်း တွေ့ရှိအရေအတွက်)

```python
df_daily_data.plot(x ='Date', y='Confirmed', kind = 'line', figsize=(10,5))
```

# Number of confirmed for each day

```python
df_diff = df_daily_data[['Confirmed', 'Lab_Neg']].diff(axis=0).fillna(0)
df_diff["Total_tests"] = df_diff.loc[:,'Confirmed':'Lab_Neg'].sum(axis = 1)
df_diff.columns = map(lambda x: 'diff_'+x, df_diff)
df_diff
```

```python
df_daily_data=df_daily_data.join(df_diff, how='outer')
df_daily_data
```
